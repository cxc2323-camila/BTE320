● Enhance your program to calculate Taxes and Net Income:

#Inventory Management
● Calculate the Economic Order Quantity using EOQ = √(2DS / H) by inputting
annual demand (D), ordering cost (S), and annual holding cost per unit (H).
○ Demand = units/year
○ Order cost = $/order
○ Holding cost = $/year/unit
● Note: have user input demand, order cost and holding cost.

```python
import math

demand = float(input("Enter annual demand (units/year): "))
order_cost = float(input("Enter ordering cost ($/order): "))
holding_cost = float(input("Enter holding cost ($/year/unit): "))

eoq = math.sqrt((2 * demand * order_cost) / holding_cost)

print(f"Economic Order Quantity (EOQ): {eoq:,.2f} units")
```

Numerical Calculator (if-elif-else statements):
● Design a simple numerical calculator in Python
● The user needs to provide three inputs:
○ Two numbers and the mathematical operator of their choice.
● The program should print the output on the screen

```python
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

if operator == "+":
    result = num1 + num2
    print("Result:", result)

elif operator == "-":
    result = num1 - num2
    print("Result:", result)

elif operator == "*":
    result = num1 * num2
    print("Result:", result)

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Cannot divide by zero")

else:
    print("Invalid operator")
```

Append items to list iteratively (while loop)
• Starting with an empty list, design a program in Python that populates the list by repeatedly asking the user for numerical inputs. The process stops when the user gives 0. Finally, the list should be displayed in descending order.

```python
numbers = []

while True:
  num = int(input('Enter a number: '))
  if num == 0:
    break
  numbers.append(num)


numbers.sort(reverse=True)
print(numbers)
```

Measures (lists)
• A quality analyst is reviewing a company's product weights to check if they are close to the desired weight. The weights are:
10.0, 9.9, 10.1, 10.05
• Initially he would like to calculate the average of these weights.
• Write a program that will calculate and display the average of a list. Use a statement to set the weights in a list:
weights = [10.0, 9.9, 10.1, 10.05]
• Example output:
Average= 10.0125
• Enhance your program to determine and display the largest value in the list.
Example output:
Average= 10.0125
Largest= 10.1

```python
weights = [10.0, 9.9, 10.1, 10.05]

total = 0
for w in weights:
  total += w

average = total / len(weights)

largest = weights[0]
for w in weights:
  if w > largest:
    largest = w

print(f"Average: {average}")
print(f"Largest: {largest}")
```

Dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else statements)
Populate an empty dictionary iteratively N times (choose your own value for N) with name:count pairs. Names (keys) are strings and counts (values) are integers. The keys must be given as inputs from the user on each iteration.
Use conditionals to check if the key has been added in the dictionary before. If not, set its count to 1; otherwise, increase its count by 1.

```python
names = {}

while True:
  name = input('Enter a name: ')

  if name == "":
    break

  if name in names:
    names[name] += 1
  else:
    names[name] = 1

print(names)
```

Classes Point and Rectangle
• Create a class Point that has:
• Two coordinate attributes (X, y)
• A method distance that calculates the distance between two instances of the class
• Create a class Rectangle that has:
• Six attributes: four Point instances and two attributes for width/height.
The latter two are calculated based on pairs of the four Point instances.
• Three methods for calculating width, height, area of the rectangle.

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(3,0)
    self.p3 = Point(0,3)
    self.p4 = Point(3,3)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height
```
