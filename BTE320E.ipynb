Parking garage (for/while loop)
• A parking garage charges $5 plus $2.50 for each hour parked.
• Write a program that prints the number of hours parked and the corresponding fee side-by-side, from 1 to 8 hours.
• Example output:
1 7.5
2 10.0
3 12.5
4 15.0
5 17.5
6 20.0
7 22.5
8 25.0
• Bonus: The minimum fee is $10 and the maximum fee is $20.

```python
print("Hour Fee")
for hour in range(1, 9):
  fee = 5 + hour * 2.5
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{hour}    {fee}")
```

Movie theater (for/while loop)
• A movie theater manager is deciding how much to spend on advertising for an upcoming movie. Each ticket is $10. Without any advertising, 20 people are forecasted to attend. Additional people may attend based on advertising spending, estimated to be:
Additional attendees = 2 x advertising Dollars 2
• In addition to advertising costs, there are fixed costs to run the theater of $200
• Write a program to generate a table for different advertising amounts and profits.
Example output (may differ slightly from your results):
Ad$ Profit
0.0 0.0
25.0 75.0
50.0 90.0
75.0 105.0
100.0 100.0
125.0 95.0
150.0 90.0
175.0 85.0
200.0 80.0

```python
ticketPrice = 10
defaultAudience = 20

print ("Ad$\tProfit")
for adMoney in range(0, 201, 25):
  revenue = (defaultAudience + 2 * round(adMoney ** 0.5)) * ticketPrice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{adMoney}\t{profit}")
```

Rock-Paper-Scissors (while loops)
• The classic hand game. Write a program that plays the game against the computer for as long as the user wants.
• Start by allowing the user to choose whether they want to play using the appropriate input.
• As long as the user chooses 'yes', the game continues.
• The rad ron i by allowing the computer to randomly choose its action, use import random # at the top of your coding cell
while ...

# ...

```python
import random

play = input("Do you want to play a game y/n?")

while play == 'y':
  player=input("Choose b/w rock, paper, scissors: ")
  computer = random.choice(['rock', 'paper', 'scissors'])

  if player == computer:
    print("Draw")
  elif player == 'rock' and computer == 'scissors':
    print("Player wins!")
  elif player == 'paper' and computer == 'rock':
    print("Player wins!")
  elif player == 'scissors' and computer == 'paper':
    print("Player wins!")
  else:
    print("Computer wins")

  play = input("Do you want to continue (y/n)? ")
```
