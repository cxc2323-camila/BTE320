Measures (lists)
• A quality analyst is reviewing a company's product weights to check if they are close to the desired weight. The weights are:
10.0, 9.9, 10.1, 10.05
• Initially he would like to calculate the average of these weights.
• Write a program that will calculate and display the average of a list. Use a statement to set the weights in a list:
weights = [10.0, 9.9, 10.1, 10.05]
• Example output:
Average= 10.0125
• Enhance your program to determine and display the largest value in the list.
Example output:
Average= 10.0125
Largest= 10.1

```python
weights = [10.0, 9.9, 10.1, 10.05]

total = 0
for w in weights:
  total += w

average = total / len(weights)

largest = weights[0]
for w in weights:
  if w > largest:
    largest = w

print(f"Average: {average}")
print(f"Largest: {largest}")
```

Alternative (simpler) method

```python
weights = [10.0, 9.9, 10.1, 10.05]

print(sum(weights) / len(weights))
print(max(weights))
```

```python
def quicksort(L):
  if len(L) < 2:
    return L
  else:
    pivot = L[len(L) // 2]
    less, equal, larger = [], [], []

    for item in L:
      if item < pivot:
        less.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        larger.append(item)
    return quicksort(less) + equal + quicksort(larger)
quicksort([8,2,5,1,6,9,7])
```

```python
def f(a,b):
  try:
    print('Now performing addition:')
    c = a + b
    print('Addition complete!\nNow performing division:')
    d = a / b
    print('Division complete!')
    return c, d
  except ZeroDivisionError:
    print("Denominator cannot be zero!")
    return float('Nan')
  except TypeError:
    print("Incompatible inputs")
    return float('Nan')
  except:
    print("Unknown error")
    return float('Nan')


f(5,0)
```
